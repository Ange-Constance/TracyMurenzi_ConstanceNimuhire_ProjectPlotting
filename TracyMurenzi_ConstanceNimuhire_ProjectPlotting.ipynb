```python
plt.figure(figsize=(8, 5))
plt.hist(df['co2'], bins=25, color='blue', edgecolor='white')
plt.title('Distribution of carbon Dioxide')
plt.show()
```

```python
#The distribution of carbon dioxide emissions is right-skewed, indicating that most vehicles emit between
#  250 and 450 grams/km, while fewer vehicles emit very high levels.
```

```python
plt.hist(df['cylinders'], bins=5, color='blue', edgecolor='white')
plt.title('Cylinders (5 bins)')
plt.xlabel('Cylinders')
plt.ylabel('Frequency')
```

```python
# This histogram shows that most vehicles have 4 to 6 cylinders, with fewer vehicles having higher cylinder counts, 
# indicating a market preference for more fuel-efficient engines and suggesting that efforts to promote sustainability 
# align well with current consumer trends.
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['highway'], bins=25, color='blue')
plt.title('Distribution of highway')
plt.show()
```

```python
# The histogram is slightly right-skewed and Most values fall between 25-40 with a peak around 28-30.
```
